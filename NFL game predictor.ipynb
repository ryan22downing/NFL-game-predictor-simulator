```python
#to randomize and simulate these stats we will use...

event = random.uniform(0,100)
time_of_possesion = abs(np.random.normal(156.16, 100, 1))

#for event = random.uniform(0,100) this creates a random number between 0 and 100 
#For time_of_possesion = abs(np.random.normal(156.16, 100, 1)) this creates a random number on a normal distribution 

#event will represent what happens to end the drive, with the numbers 0-100 mapping to the probabilities shown above
#time_of_possesion will represent how long a drive takes, with these numbers being made from a random sample of 50 drives 

```

```python
#This simulates a single game and explains what happens each drive
#We use the time and event randomizers to run through a loop in which every drive has a randomized time and outcome
#This program is looped until total time reaches 3600, which is the amount of seconds in an NFL game
#The final score is printed at the end, but you can also see the evolution of the score as drives are completed by each team


points = []


points2 = []



time = 0
count = 0 

while time < 3600:
    event = random.uniform(0,100)
    time_of_possesion = abs(np.random.normal(156.16, 100, 1))
    if event >= 0 and event < 42.494:
        print("Team 1: punts")
    if event >= 42.494 and event < 63.837:
        points.append(6)
        extra = random.uniform(0,100)
        if extra >=0 and extra < 85:
            print("Team 1: Touchdown, and converts the extra point")
            points.append(1)
        if extra >= 85 and extra < 90:
            print("Team 1: Touchdown, and converts the 2pt conversion")
            points.append(2)
        if extra >= 90 and extra < 100:
            print("Team 1: Touchdown, failed extra point attempt")
    if event >= 63.837 and event < 78.372:
        print("Team 1: Field Goal")
        points.append(3)
    if event >= 78.372 and event < 86.939:
        print("Team 1: Throws Interception to Team 2")
    if event >= 86.939 and event < 92.259:
        print("Team 1: Fumbles the ball, recovered by Team 2")
    if event >= 92.259 and event < 96.507:
        print("Team 1: Turnover on downs")
    if event >= 96.507 and event < 99.364:
        print("Team 1: Missed FG")
    if event >= 99.364 and event < 99.642:
        print("Team 1: Tackled in the end zone for a safety")
        points2.append(2)
    if event >= 99.642 and event < 99.840:
        print("Team 1: Punt blocked by Team 2")
    if event >= 99.840 and event < 100:
        print("Team 1: FG blocked by Team 2")
    print(sum(points), sum(points2))


    event2 = random.uniform(0,100)
    time_of_possesion2 = abs(np.random.normal(156.16, 100, 1))
    if time + time_of_possesion + time_of_possesion2 > 3600:
        break
    if event2 >= 0 and event2 < 42.494:
        print("Team 2: punts")
    if event2 >= 42.494 and event2 < 63.837:
        points2.append(6)
        extra = random.uniform(0,100)
        if extra >=0 and extra < 85:
            print("Team 2: Touchdown, and converts the extra point")
            points2.append(1)
        if extra >= 85 and extra < 90:
            print("Team 2: Touchdown, and converts the 2pt conversion")
            points2.append(2)
        if extra >= 90 and extra < 100:
            print("Team 2: Touchdown, failed extra point attempt")
    if event2 >= 63.837 and event2 < 78.372:
        print("Team 2: Field Goal")
        points2.append(3)
    if event2 >= 78.372 and event2 < 86.939:
        print("Team 2: Throws Interception to Team 1")
    if event2 >= 86.939 and event < 92.259:
        print("Team 2: Fumbles the ball, recovered by Team 1")
    if event2 >= 92.259 and event2 < 96.507:
        print("Team 2: Turnover on downs")
    if event2 >= 96.507 and event2 < 99.364:
        print("Team 2: Missed FG")
    if event2 >= 99.364 and event2 < 99.642:
        print("Team 2: Tackled in the end zone for a safety")
        points.append(2)
    if event2 >= 99.642 and event2 < 99.840:
        print("Team 2: Punt blocked by Team 1")
    if event2 >= 99.840 and event2 < 100:
        print("Team 2: FG blocked by Team 1")
    print(sum(points), sum(points2))  
    time += (time_of_possesion+time_of_possesion2)
print("Final Score:", "Team 1:", sum(points), "Team 2:", sum(points2))

#below is an example output
```

```python
#This code has the ability to simulate many games, then convert it to a new database with x amount of scores and various stats 
#This is useful for creating betting lines, as percentages of events occuring can be changed team by team 
#You can simulate 10,000 or more games of two specific teams and get a mean score outcome of the game, or any other stat...
count = 0


scores = pd.DataFrame(columns=['team1', 'team2', 'pointdif', 'totalpoints', 'punts team 1', 'punts team 2', 'turnoverstm1', 'turnoverstm2', ])


while count < 10000:
    points = []
    points2 = []

    punt = []
    punt2 = []

    interception = []
    interception2 = []

    fumble = []
    fumble2 = []

    turnoverondowns = []
    turnoverondowns2 = []

    missedfg = []
    missedfg2 = []

    blockedpunt = []
    blockedpunt2 = []

    blockedfg = []
    blockedfg2 = []


    time = 0

    
    
    while time < 3600:
        event = random.uniform(0,100)
        time_of_possesion = abs(np.random.normal(156.16, 100, 1))
        if event >= 0 and event < 42.494:
            punt.append(1)
        if event >= 42.494 and event < 63.837:
            points.append(6)
            extra = random.uniform(0,100)
            if extra >=0 and extra < 85:
                points.append(1)
            if extra >= 85 and extra < 90:
                points.append(2)
        if event >= 63.837 and event < 78.372:
            points.append(3)
        if event >= 78.372 and event < 86.939:
            interception.append(1)
        if event >= 86.939 and event < 92.259:
            fumble.append(1)
        if event >= 92.259 and event < 96.507:
            turnoverondowns.append(1)
        if event >= 96.507 and event < 99.364:
            missedfg.append(1)
        if event >= 99.364 and event < 99.642:
            points.append(2)
        if event >= 99.642 and event < 99.840:
            blockedpunt.append(1)
        if event >= 99.840 and event < 100:
            blockedfg.append(1)


        event2 = random.uniform(0,100)
        time_of_possesion2 = abs(np.random.normal(156.16, 100, 1))
        if event2 >= 0 and event2 < 42.494:
            punt2.append(1)
        if event2 >= 42.494 and event2 < 63.837:
            points2.append(6)
            extra = random.uniform(0,100)
            if extra >=0 and extra < 85:
                points2.append(1)
            if extra >= 85 and extra < 90:
                points2.append(2)
        if event2 >= 63.837 and event2 < 78.372:
            points2.append(3)
        if event2 >= 78.372 and event2 < 86.939:
            interception2.append(1)
        if event2 >= 86.939 and event < 92.259:
            fumble2.append(1)
        if event2 >= 92.259 and event2 < 96.507:
            turnoverondowns2.append(1)
        if event2 >= 96.507 and event2 < 99.364:
            missedfg2.append(1)
        if event2 >= 99.364 and event2 < 99.642:
            points2.append(2)
        if event2 >= 99.642 and event2 < 99.840:
            blockedpunt2.append(1)
        if event2 >= 99.840 and event2 < 100:
            blockedfg2.append(1)
 
        time += (time_of_possesion+time_of_possesion2)
    score = sum(points)
    score2 = sum(points2)
    punts = sum(punt)
    punts2 = sum(punt2)
    turnovers = (sum(interception) + sum(fumble))
    turnovers2 = (sum(interception2) + sum(fumble2))
    pointdif = (score - score2)
    pointcomb = (score + score2)
    
    count += 1
    scores = scores.append({'team1': score, 'team2': score2, 'pointdif': pointdif, 'totalpoints': pointcomb, 'punts team 1': punts, 'punts team 2': punts2, 'turnoverstm1': turnovers, 'turnoverstm2': turnovers2}, ignore_index=True)


print(scores)

#below is an example output
```
